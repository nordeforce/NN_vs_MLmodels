```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

table = in_data

print("Fitur (table.X) shape:", table.X.shape)
print("Target (table.Y) shape:", table.Y.shape)
print("Fitur names:", [v.name for v in table.domain.attributes])
print("Target names:", [v.name for v in table.domain.class_vars])

# Tampilkan nama kolom yang terbaca
print("Kolom domain:", [v.name for v in table.domain.variables])

# Ambil indeks kolom fitur
idx_e  = table.domain.index("e")
idx_de = table.domain.index("de")

# Ambil nilai fitur
x = table.X[:, idx_e]
y = table.X[:, idx_de]

# Ambil nilai pwm sebagai target (1-dim array)
z = table.Y.flatten()  # langsung array 1 dimensi

# Buat plot 3D
fig = plt.figure(figsize=(7, 5))
ax  = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='blue', marker='o', alpha=0.6)
sc = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o', alpha=0.8)
fig.colorbar(sc, ax=ax, label="pwm")

ax.set_xlabel("e")
ax.set_ylabel("de")
ax.set_zlabel("pwm")

ax.set_title("3D Scatter NN: pwm vs e & de")
plt.show()
```
